```
import numpy as np
import cv2
```

```
from google.colab.patches import cv2_imshow
```

SIMPLE THRESHOLDING- For every pixel, the same threshold value is applied. If the pixel value is smaller than the threshold, it is set to 0, otherwise it is set to a maximum value. 

```
#Read image
img=cv2.imread("gardient.jpg",0)
cv2_imshow(img)
```

```
#Converting bgr image to rgb image
imgg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2_imshow(imgg)
```

```
#thresholding
_,thresh1 = cv2.threshold(imgg,127,255,cv2.THRESH_BINARY)
cv2_imshow(thresh1)
```

```
#Inverse of binary thresholding
_,thresh2 = cv2.threshold(imgg,50,255,cv2.THRESH_BINARY_INV)
cv2_imshow(thresh2)
```

```
_,thres3=cv2.threshold(imgg,127,255,cv2.THRESH_TRUNC)
cv2_imshow(thres3)
```

```
_,thres4=cv2.threshold(imgg,127,255,cv2.THRESH_TOZERO)
cv2_imshow(thres4)
```

```
_,thres5=cv2.threshold(imgg,127,255,cv2.THRESH_TOZERO_INV)
cv2_imshow(thres5)
```

ADAPTIVE THRESHOLDING

Adaptive thresholding is the method where the threshold value is calculated for smaller regions and therefore, there will be different threshold values for different regions.

```
#Read image 
img2=cv2.imread("s.png",0)
#convert to rgb
img3=cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
cv2_imshow(img3)
```

ADAPTIVE_THRESH_MEAN_C - The threshold value is the mean of the neighbourhood area minus the constant C

```
graying = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
thre=cv2.adaptiveThreshold(graying,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
cv2_imshow(thre)
```

cv2.ADAPTIVE_THRESH_GAUSSIAN_C -- The threshold value is a gaussian-weighted sum of the neighbourhood values minus the constant C.

```
th3 = cv2.adaptiveThreshold(graying,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY, 11, 2)
cv2_imshow(th3)
```
