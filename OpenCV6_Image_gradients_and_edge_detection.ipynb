```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```
img=cv2.imread("testing.jpg",cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)
```

LAPLACIAN GRADIENT

```
lap=cv2.Laplacian(img,cv2.CV_64F,ksize=3)
```

```
lap=np.uint8(np.absolute(lap))
```

```
cv2_imshow(lap)
```

SOBEL GRADIENT

```
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0)
sobely=cv2.Sobel(img,cv2.CV_64F,0,1)
```

```
sobelx = np.uint8(np.absolute(sobelx))
sobely = np.uint8(np.absolute(sobely))
```

```
sobcomb=cv2.bitwise_or(sobelx,sobely)
```

```
cv2_imshow(sobelx)
```

```
cv2_imshow(sobely)
```

```
cv2_imshow(sobcomb)
```

CANNY EDGE DETECTOR

```
can=cv2.Canny(img,100,200)
cv2_imshow(can)
```

```
titles = ['image', 'Laplacian','sobelx','sobely','sobel combined','canny']
images = [img, lap, sobelx, sobely, sobcomb,can]
```

```
#Using matplotlib
for i in range(6):
    plt.subplot(2,3, i+1), plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
