```
import cv2
import numpy as np
```

```
from google.colab.patches import cv2_imshow
```

```
img=cv2.imread("Lenna.png")
cv2_imshow(img)
```

IMAGE PYRAMIDS

```
py1=cv2.pyrDown(img)
cv2_imshow(py1)
```

```
py2=cv2.pyrDown(py1)
cv2_imshow(py2)
```

```
py3=cv2.pyrUp(py2)
cv2_imshow(py3)
```

Using Gaussian pyramid

```
img2=img.copy()
gp=[img2]
```

```
#creating image pyramid using pyrDown
for i in range(6):
  img2=cv2.pyrDown(img2)
  gp.append(img2)
  cv2_imshow(img2)
```

```
lay=gp[5]
cv2_imshow(lay)
```

LAPLACIAN PYRAMID-- A Laplacian pyramid is very similar to a Gaussian pyramid but saves the difference image of the blurred versions between each levels.

```
lp=[lay]
for i in range(5,0,-1):
  gaus=cv2.pyrUp(gp[i])
  lap=cv2.subtract(gp[i-1].shape[1],gaus.shape[0])
  cv2_imshow(lap)
```
