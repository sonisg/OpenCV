```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```
img=cv2.imread("shapess.png")
cv2_imshow(img)
```

HARRIS CORNER DETECTOR

```
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)
dest = cv2.cornerHarris(gray,2,3,0.04)
```

```
dest = cv2.dilate(dest,None)
```

```
img[dest > 0.01 * dest.max()] = [0, 0, 255]
cv2_imshow(img)
```

SRI TOMASI CORNER DETECTOR

```
img1=cv2.imread("shapess.png")
cv2_imshow(img1)
```

```
grey = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
corners=cv2.goodFeaturesToTrack(grey,225,0.01,10)
corners=np.int0(corners)
```

```
for i in corners:
    x,y = i.ravel()
    cv2.circle(img1, (x, y),3,[255, 255, 0],-1)
```

```
cv2_imshow(img1)
```
