```
import numpy as np
```

BITWISE OPERATION ON IMAGES

```
#Adding blank image
img3=np.zeros((250,500,3),np.uint8)
img3=cv2.rectangle(img3,(200,0),(300,100),(255,255,255),-1)
cv2_imshow(img3)
print(img3.shape)
```

```
img4=cv2.imread("image1.jpg")
img4=cv2.resize(img4,(500,250))
cv2_imshow(img4)
```

```
bitAnd = cv2.bitwise_and(img3, img4)
bitOr = cv2.bitwise_or(img3, img4)
bitXor = cv2.bitwise_xor(img3, img4)
bitNot1 = cv2.bitwise_not(img3)
bitNot2 = cv2.bitwise_not(img4)
```

```
cv2_imshow(bitAnd)
```

```
cv2_imshow(bitOr)
```

```
cv2_imshow(bitXor)
```

```
cv2_imshow( bitNot1)
```

```
cv2_imshow(bitNot2)
```
