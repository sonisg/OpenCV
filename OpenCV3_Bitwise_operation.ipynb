BITWISE OPERATION ON IMAGES

```
img3=cv2.imread("1.png")
cv2_imshow(img3)
```

```
img4=cv2.imread("2bit2.png")
cv2_imshow(img4)
```

```
bitAnd = cv2.bitwise_and(img3, img4)
bitOr = cv2.bitwise_or(img3, img4)
bitXor = cv2.bitwise_xor(img3, img4)
bitNot1 = cv2.bitwise_not(img3)
bitNot2 = cv2.bitwise_not(img4)
```

```
cv2_imshow(bitAnd)
```

```
cv2_imshow( bitOr)
```

```
cv2_imshow(bitXor)
```

```
cv2_imshow( bitNot1)
```

```
cv2_imshow(bitNot2)
```
