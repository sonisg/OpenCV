```
import cv2
import matplotlib.pyplot as plt
```

```
from google.colab.patches import cv2_imshow
```

```
# use 1 for color image
img = cv2.imread ("tesst.jpg",1)
```

```
# use 0 for black white image
imgg=cv2.imread('tesst.jpg',0)
```

```
#show image using imshow
cv2_imshow(img)
```

```
#show image using imshow
cv2_imshow(imgg)
```

```
#EXtracting region of interest
roi = imgg[100 : 500, 200 : 700] 
cv2_imshow(roi)
```

```
img = cv2.imread('tesst.jpg', cv2.IMREAD_COLOR)
# If the image path is wrong, the resulting img will be none
if img is None:
    print('Open Error')
else:
    print('Image Loaded')
```

```
#resize image using resize
res=cv2.resize(imgg,(600,400))
cv2_imshow(res)
```

```
#prints the shape of the image
print(img.shape)
```

```
#prints size of the image
print(img.size)
```

```
#prints type of image
print(img.dtype)
```

```
#gives blue green red channels
b,g,r=cv2.split(img)
```

```
print(b)
```

```
#merge the image 
gimg=cv2.merge([b,g,r])
cv2_imshow(gimg)
```
